```python
import pandas as pd
```

```python
dataset = pd.read_excel('./datasets/izborni_predmeti.xlsx')
dataset
```

```python
f = lambda x: x['NAZIV_SMERA'] in ['Informatika', 'Racunarstvo i informatika']
dataset = dataset[dataset.apply(f, axis=1)]
```

```python
import numpy as np
```

```python
dataset = dataset.groupby(['INDEKS'])['NAZIV_PREDMETA'].apply(list).values.tolist()
dataset
```

```python
# Apriori za min sup 2% in min coef 80%
from apriori_python import apriori
```

```python
freqItem, rules = apriori(dataset, minSup=0.02, minConf=0.80)
```
