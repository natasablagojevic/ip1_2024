```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
```

```python
data = pd.read_csv('../datasets/data.csv')
```

```python
data
```

```python
data.isna().any()
```

```python
data.isna().sum()
```

```python
data.shape
```

```python
data.drop('Unnamed: 32', axis=1, inplace=True)
```

```python
data.drop('id', axis=1, inplace=True)
```

```python
data
```

```python
Y = data['diagnosis']
X = data.drop('diagnosis', axis=1)
```

```python
X
```

```python
feature_names = X.columns
```

```python
feature_names
```

```python
Y.unique()
```

```python
lines, edges, bins = plt.hist(Y)
plt.bar_label(bins)
plt.show()
```

```python
Y.replace('B', 0, inplace=True)
Y.replace('M', 1, inplace=True)
```

```python
Y.unique()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=123, stratify=Y)
```

```python
X_train.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
def report(y, y_pred):
    print(classification_report(y, y_pred))
    print('-'*100)
    print(pd.DataFrame(confusion_matrix(y, y_pred), columns=['B', 'M'], index=['B', 'M']))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
report(y_train, model.predict(X_train))
```

```python
report(y_test, model.predict(X_test))
```

```python
params = {
    'n_neighbors': range(10,50,10),
    'weights': ['uniform', 'distance'],
    'p': [1,2]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
estimator = GridSearchCV(KNeighborsClassifier(), param_grid=params, cv=5, verbose=5)
```

```python
estimator.fit(X_train, y_train)
```

```python
estimator.best_params_
```

```python
model = estimator.best_estimator_
```

```python
report(y_train, model.predict(X_train))
```

```python
report(y_test, model.predict(X_test))
```
