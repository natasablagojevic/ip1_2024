```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.cluster import KMeans
```

```python
dogs = pd.read_csv('../datasets/dogs.csv')
```

```python
dogs
```

```python
feature_names = dogs.columns[1:]
```

```python
X = dogs[feature_names]
```

```python
X 
```

```python
plt.scatter(X['height'], X['weight'])
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X = scaler.fit_transform(X)
```

```python
X
```

```python
X = pd.DataFrame(X, columns=feature_names)
```

```python
X
```

```python
kmeans = KMeans(n_clusters=2, n_init='auto')
```

```python
kmeans.fit(X)
```

```python
kmeans.labels_
```

```python
X[kmeans.labels_ == 0]
```

```python
X[kmeans.labels_ == 1]
```

```python
centers = pd.DataFrame(kmeans.cluster_centers_, columns=feature_names)
```

```python
centers
```

```python
plt.scatter(centers['height'], centers['weight'], marker='X', color='black')

for label in np.unique(kmeans.labels_):
    elems = X[kmeans.labels_ == label]
    plt.scatter(elems['height'], elems['weight'], label=label)
    
plt.xlabel('height')
plt.ylabel('weight')
plt.title('KMeans')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(X, kmeans.labels_)
```

```python
from sklearn.metrics import silhouette_samples
```

```python
silhouette_values = silhouette_samples(X, kmeans.labels_)
```

```python
silhouette_values
```

```python
plt.scatter(X['height'], X['weight'], c=silhouette_values, cmap='Greens')
plt.colorbar()
plt.title('Silhouette coefs')
plt.show()
```

```python
# IZBOR BROJA KLASTERA
```

```python
ks = range(2, X.shape[0])
inits = ['random', 'k-means++']
fig = plt.figure(figsize=(10, 30))
idx = 1 
silhouette = []
inertias = []

for k in ks:
    for init in inits:
        kmeans = KMeans(n_clusters=k, init=init, n_init='auto')
        kmeans.fit(X)
        
        if init == 'k-means++':
            inertias.append(kmeans.inertia_)
            silhouette.append(silhouette_score(X, kmeans.labels_))
            
        fig.add_subplot(len(ks), len(inits), idx)
        idx += 1 
        
        for label in kmeans.labels_:
            elems = dogs[kmeans.labels_ == label]
            plt.scatter(elems['height'], elems['weight'])
            
        centroids = pd.DataFrame(scaler.inverse_transform(kmeans.cluster_centers_), columns=feature_names)
        plt.scatter(centroids['height'], centroids['weight'], color='black', marker='X')
        
        plt.title(f'k={k}, init={init}, inertia={kmeans.inertia_}')
        
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(ks, inertias, marker='o')
plt.xlabel('num_of_clusters')
plt.ylabel('inerta')
plt.title('Elbow method')

plt.subplot(1, 2, 2)
plt.plot(ks, silhouette, marker='o')
plt.xlabel('num_of_clusters')
plt.ylabel('silhouette scores')
plt.title('Silhouette coefs')

plt.show()
```

```python
from matplotlib.image import imread
```

```python
img = imread('../datasets/masa.jpg')
```

```python
img_size = img.shape
```

```python
img_size
```

```python
X_img = img.reshape(img_size[0] * img_size[1], img_size[2])
```

```python
kmeans = KMeans(n_clusters=30, n_init='auto')
```

```python
kmeans.fit(X_img)
```

```python
X_comp = kmeans.cluster_centers_[kmeans.labels_]
X_comp = np.clip(X_comp.astype('uint8'), 0, 255)
```

```python
X_comp = X_comp.reshape(img_size[0], img_size[1], img_size[2])
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 8))
ax[0].imshow(img)
ax[0].set_title('Original')
ax[1].imshow(X_comp)
ax[1].set_title('Compressed')

for ax in fig.axes:
    ax.axis('off')
    
plt.tight_layout()
```

```python
# Bisecting KMeans
```

```python
from sklearn.cluster import BisectingKMeans
```

```python
bkmeans = BisectingKMeans(n_clusters=3, bisecting_strategy='largest_cluster')
```

```python
bkmeans.fit(X)
```

```python
def visualize(data, centers, labels, feature_names, alg_name):
    plt.scatter(centers[:,0], centers[:,1], marker='X', label='centroids', color='black')
    
    for label in np.unique(bkmeans.labels_):
        elems = data[bkmeans.labels_ == label]
        plt.scatter(elems[feature_names[0]], elems[feature_names[1]], label=label)
        
    plt.xlabel(feature_names[0])
    plt.ylabel(feature_names[1])
    plt.title(f'{alg_name} {len(centers)} clusters')
    plt.legend()
```

```python
visualize(X, bkmeans.cluster_centers_, bkmeans.labels_, feature_names, 'BisectingKMeans')
```
