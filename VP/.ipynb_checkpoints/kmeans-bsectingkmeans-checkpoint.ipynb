```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import MinMaxScaler
```

```python
dogs = pd.read_csv('../datasets/dogs.csv')
```

```python
dogs
```

```python
plt.scatter(dogs['height'], dogs['weight'])
plt.xlabel('height')
plt.ylabel('weigth')
plt.title('Dogs')
plt.show()
```

```python
feature_names = dogs.columns[1:]
feature_names
```

```python
X = dogs[feature_names]
```

```python
X
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X)
```

```python
X = scaler.transform(X)
```

```python
X
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=2, n_init='auto')
```

```python
model.fit(X)
```

```python
model.labels_
```

```python
dogs[model.labels_ == 0]
```

```python
dogs[model.labels_ == 1]
```

```python
model.cluster_centers_
```

```python
centers = pd.DataFrame(model.cluster_centers_, columns=feature_names)
```

```python
centers
```

```python
model.inertia_
```

```python
X = pd.DataFrame(X, columns=feature_names)
```

```python
X
```

```python
plt.scatter(centers['height'], centers['weight'], marker='X', 
            label='centroids')

for c in np.unique(model.labels_):
    elems = X[model.labels_ == c]
    plt.scatter(elems['height'], elems['weight'], label=c)
    
plt.xlabel('height')
plt.ylabel('weight')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import silhouette_score
```

```python
ks = range(2, X.shape[0])
inits = ['random', 'k-means++']
fig = plt.figure(figsize=(10, 30))
idx = 1 
silhouette = []
inertias = []

for k in ks:
    for init in inits:
        kmeans = KMeans(n_clusters=k, n_init='auto', init=init)
        kmeans.fit(X)
#         silhouette.append(silhouette_score(X, kmeans.labels_))
        
        if init == 'k-means++':
            inertias.append(kmeans.inertia_)
            silhouette.append(silhouette_score(X, kmeans.labels_))
            
        fig.add_subplot(len(ks), len(inits), idx)
        idx += 1
        
        for label in range(k):
            cluster = dogs[kmeans.labels_ == c]
            plt.scatter(cluster['height'], cluster['weight'])
            
        centroids = pd.DataFrame(scaler.inverse_transform(kmeans.cluster_centers_), columns=feature_names)
        plt.scatter(centroids['height'], centroids['weight'], color='black', marker='X')
        
        plt.title(f'k={k}, init={init}, inertia={kmeans.inertia_}')
        
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(ks, inertias, marker='o')
plt.xlabel('num_of_cluster')
plt.ylabel('inertia')
plt.title('Elbow method')

plt.subplot(1, 2, 2)
plt.plot(ks, silhouette, marker='o')
plt.xlabel('num_of_cluster')
plt.ylabel('silhouette coefs')
plt.title('Silhouette score')

plt.show()
```

```python
from sklearn.cluster import BisectingKMeans
```

```python
bkmeans = BisectingKMeans(n_clusters=3, bisecting_strategy='largest_cluster')
```

```python
bkmeans
```

```python
def visualize(data, centers, labels, feature_names, alg_name):
    plt.scatter(centers[:,0], centers[:,1], marker='X', label='centers', color='black')
    
    for c in np.unique(labels):
        elems = data[labels == c]
        plt.scatter(elems[feature_names[0]], elems[feature_names[1]], label=c)
        
    plt.xlabel(feature_names[0])
    plt.ylabel(feature_names[1])
    plt.title(f'{alg_name} {len(centers)}')
    plt.legend()
    plt.show()
```

```python
bkmeans.fit(X)
```

```python
visualize(X, bkmeans.cluster_centers_, bkmeans.labels_, feature_names, 'BisectingKMeans')
```
