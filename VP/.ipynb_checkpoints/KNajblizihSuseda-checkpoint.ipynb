```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
```

```python
data = pd.read_csv('../datasets/data.csv')
```

```python
data
```

```python
data.describe()
```

```python
data.isna().any()
```

```python
data.isna().sum()
```

```python
data.shape
```

```python
data.drop('Unnamed: 32', axis=1, inplace=True)
```

```python
data.drop('id', axis=1, inplace=True)
```

```python
y = data['diagnosis']
X = data.drop('diagnosis', axis=1)
```

```python
feature_names = X.columns
```

```python
y.hist()
```

```python
lines, edges, bins = plt.hist(y, bins=2)
plt.bar_label(bins)
```

```python
y.replace('B', 0, inplace=True)
y.replace('M', 1, inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=456, stratify=y)
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
model = KNeighborsClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
def report(y, y_pred):
    print(classification_report(y, y_pred))
    print('-'*100)
    print(pd.DataFrame(confusion_matrix(y, y_pred), columns=['B', 'M'], index=['B', 'M']))
```

```python
report(y_train, model.predict(X_train))
```

```python
report(y_test, model.predict(X_test))
```

```python
model.get_params()
```

```python
params = {
    'p': [1,2],
    'n_neighbors': range(10,50,5),
    'weights': ['uniorm', 'distance']
}
```

```python
estimator = GridSearchCV(KNeighborsClassifier(), param_grid=params, verbose=5, cv=9)
```

```python
estimator.fit(X_train, y_train)
```

```python
estimator.best_params_
```

```python
model = estimator.best_estimator_
```

```python
report(y_train, model.predict(X_train))
```

```python
report(y_test, model.predict(X_test))
```
