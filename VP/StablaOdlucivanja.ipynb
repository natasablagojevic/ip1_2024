```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.tree import DecisionTreeClassifier
```

```python
data = pd.read_csv('../datasets/data.csv')
```

```python
data
```

```python
data.isna().any()
```

```python
data.isna().sum()
```

```python
data.shape
```

```python
data.drop('Unnamed: 32', axis=1, inplace=True)
```

```python
data.drop('id', axis=1, inplace=True)
```

```python
data
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
def report(y, y_pred):
    print(classification_report(y, y_pred))
    print('-'*80)
    print(pd.DataFrame(confusion_matrix(y, y_pred), columns=['B', 'M'], index=['B', 'M']))
```

```python
y = data['diagnosis']
X = data.drop('diagnosis', axis=1)
```

```python
feature_names = X.columns
```

```python
y.replace('B', 0, inplace=True)
y.replace('M', 1, inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=123, stratify=y)
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
report(y_train, model.predict(X_train))
```

```python
report(y_test, model.predict(X_test))
```

```python
model.get_depth()
```

```python
model.get_n_leaves()
```

```python
model.get_params()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [2,4,6,8,10]
}
```

```python
estimator = GridSearchCV(DecisionTreeClassifier(), param_grid=params)
```

```python
estimator.fit(X_train, y_train)
```

```python
estimator.best_estimator_
```

```python
model = estimator.best_estimator_
```

```python
report(y_train, model.predict(X_train))
```

```python
report(y_test, model.predict(X_test))
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(20,30))
plot_tree(model, max_depth= model.get_depth(),feature_names=feature_names, filled=True)
plt.title('Decision tree classifier')
plt.show()
```
