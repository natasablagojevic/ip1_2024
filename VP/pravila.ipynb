```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
data = pd.read_excel('../datasets/izborni_predmeti.xlsx')
```

```python
data
```

```python
# filtriranje studenata I i R smera
f = lambda x: x['NAZIV_SMERA'] in ['Informatika', 'Racunarstvo i informatika']
```

```python
data = data[data.apply(f, axis=1)]
```

```python
data
```

```python
# grupisanje po indeksu

data = data.groupby(['INDEKS'])['NAZIV_PREDMETA'].apply(list).values.tolist()
data
```

```python
freq, rules = apriori(data, minSup=0.02, minConf=0.8)
```

```python
pd.DataFrame(rules, columns=['Antecedent', 'Consequent', 'Confidence'])
```

```python
# Consequent = skup predmeta koji se pojavljuju na desnoj strani pravila (glava)
# Antecedent = skup predmeta koji se pojavljuju na levoj strani pravila (telo)
# Confidence = verovatnoca da ce se consequent pojaviti ako je antecedent prisutan
```
