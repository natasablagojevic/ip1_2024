```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../datasets/balloons.csv')
```

```python
df.head()
```

```python
df.isna().any()
```

```python
df.dropna(axis=0, inplace=True)
```

```python
df.isna().any()
```

```python
Y = df['inflated']
X = df.drop('inflated', axis=1)
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, stratify=Y, random_state=123)
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
encoder = OrdinalEncoder()
```

```python
encoder.fit(X_train)
```

```python
X_train = encoder.transform(X_train)
```

```python
X_test = encoder.transform(X_test)
```

```python
from sklearn.naive_bayes import CategoricalNB
```

```python
model = CategoricalNB()
```

```python
model.fit(X_train, y_train)
```

```python
model.category_count_
```

```python
model.class_count_
```

```python
y_train_pred = model.predict(X_train)
```

```python
y_test_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_train, y_train_pred)
```

```python
confusion_matrix(y_test, y_test_pred)
```

```python
# KLASIFIKACIJA TEKSTA

from sklearn.feature_extraction import DictVectorizer

dv = DictVectorizer()
dv.fit(X_train)

sparse_matrix = dv.transofrm(X_train)

X_train = pd.DataFrame(sparse_matrix.toarray(), columns=dv.feature_names_)

X_train.head()

from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()

model.fit(X_train, y_train)

# broj pojavljivanja svake od klasa
model.class_count_

# broj pojavljivanja svake od reci u svakoj od klasa
model.feature_count_

y_train_pred = model.predict(X_train)

confusion_matrix(y_train, y_train_pred)

from sklearn.metrics import accuracy_score

accuracy_score(y_train, y_train_pred)

from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

confusion_matrix(y_test, y_pred)
```
