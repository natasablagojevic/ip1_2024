```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../datasets/data.csv')
```

```python
data
```

```python
data.isna().any()
```

```python
data.isna().sum()
```

```python
data.shape
```

```python
data.drop('Unnamed: 32', axis=1, inplace=True)
```

```python
data.drop('id', axis=1, inplace=True)
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
```

```python
Y = data['diagnosis']
X = data.drop('diagnosis', axis=1)
```

```python
edge, lines, bins = plt.hist(Y)
plt.bar_label(bins)
plt.show()
```

```python
Y.replace('B', 0, inplace=True)
```

```python
Y.replace('M', 1, inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=123, stratify=Y)
```

```python
X_train.shape
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
def report(y, y_pred):
    print(classification_report(y, y_pred))
    print('-'*80)
    print(pd.DataFrame(confusion_matrix(y, y_pred), index=['B', 'M'], columns=['B', 'M']))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
model.get_params()
```

```python
report(y_train, model.predict(X_train))
```

```python
report(y_test, model.predict(X_test))
```

```python
from sklearn.tree import plot_tree
```

```python
feature_names = X.columns
```

```python
plt.figure(figsize=(12,8))
plot_tree(model, feature_names=feature_names, filled=True, class_names=True, max_depth=3)
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
model.get_params()
```

```python
params = {
    'max_depth': [2,4,6,8,10],
    'criterion': ['gini', 'entropy']
}
```

```python
estimator = GridSearchCV(DecisionTreeClassifier(), param_grid=params, cv=5, verbose=5)
```

```python
estimator.fit(X_train, y_train)
```

```python
estimator.best_estimator_
```

```python
model = estimator.best_estimator_
```

```python
report(y_train, model.predict(X_train))
```

```python
report(y_test, model.predict(X_test))
```

```python
plt.figure(figsize=(12,8))
plot_tree(model, feature_names=feature_names, filled=True)
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
forest = RandomForestClassifier(n_estimators=2)
```

```python
forest.fit(X_train, y_train)
```

```python
report(y_test, forest.predict(X_test))
```
