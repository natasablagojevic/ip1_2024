```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../datasets/dogs.csv')
```

```python
feature_names = df.columns[1:]
```

```python
X = df[feature_names]
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X = scaler.fit_transform(X)
```

```python
X = pd.DataFrame(X, columns=feature_names)
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model = AgglomerativeClustering(n_clusters=3, linkage='single', compute_distances=True)
```

```python
model.fit(X)
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
Z = linkage(X, method='single')
```

```python
from sklearn.metrics import silhouette_score
```

```python
k = 3
linkages = ['single', 'complete', 'average']
fig = plt.figure(figsize=(15, 5))

for idx, linkage in enumerate(linkages):
    model = AgglomerativeClustering(n_clusters=k, linkage=linkage)
    model.fit(X)
    
    fig.add_subplot(1, len(linkages), idx+1)
    
    for c in range(k):
        elems = df[model.labels_ == c]
        plt.scatter(elems['height'], elems['weight'])
    
    plt.title(f'l={linkage}, sil={silhouette_score(X, model.labels_)}')
```

```python
min_samples = 2
eps_values = [0.1, 0.15, 0.2, 0.25, 0.3]
min_samples_values = range(2, 5)
fig = plt.figure(figsize=(25,10))
idx = 1

for eps in eps_values:
    for min_samples in min_samples_values:
        model = DBSCAN(eps=eps, min_samples=min_samples)
        model.fit(X)
        
        fig.add_subplot(len(eps_values), len(min_samples_values), idx)
        idx += 1
        
        for cluster_label in np.unique(model.labels_):
            elems = df[model.labels_ == cluster_label]
            if cluster_label == -1:
                plt.scatter(elems['height'], elems['weight'], color='black', label='sum')
            else:
                plt.scatter(elems['height'], elems['weight'], label=f'klaster {cluster_label}')
                
        plt.legend()
        
        if len(np.unique(model.labels_)) == 1:
            s = None 
        else:
            s = silhouette_score(X, model.labels_)
            
        plt.title(f'eps={eps}, m={min_samples}, sil={s}')
        
plt.tight_layout()
```
