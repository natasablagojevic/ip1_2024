```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
```

```python
data
```

```python
X = data['data']
Y = data['target']
```

```python
feature_names = data['feature_names']
```

```python
class_names = data['target_names']
```

```python
X.shape
```

```python
Y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=123, stratify=Y)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(X_train)
```

```python
pca.components_.shape
```

```python
pca.components_
```

```python
plt.matshow(pca.components_)
plt.xticks(ticks=range(pca.components_.shape[1]), labels=feature_names, rotation='vertical')
plt.yticks(ticks=range(pca.n_components_), labels=pca.get_feature_names_out())
plt.colorbar()
plt.show()
```

```python
pca.explained_variance_
```

```python
pca.explained_variance_ratio_
```

```python
plt.bar(x=pca.get_feature_names_out(), height=pca.explained_variance_ratio_)
plt.title('Udeo objasnjene varijanse svake od glavnih komponenti pojedinacno')
plt.show()
```

```python
cumulative_variance_ratio = np.cumsum(pca.explained_variance_ratio_)
```

```python
plt.plot(pca.get_feature_names_out(), cumulative_variance_ratio, marker='o')
plt.title('Kumulativna suma udela objasnjene varijanse')
plt.show()
```

```python
X_train_pca = pca.transform(X_train)
X_test_pa = pca.transform(X_test)
```

```python
X_train_pca.shape
```

```python
scatter = plt.scatter(X_train_pca[:, 0], X_train_pca[:, 1], c=y_train)
plt.xlabel('First principal component')
plt.ylabel('Second principal component')
plt.legend(handles=scatter.legend_elements()[0], labels=class_names.tolist())
plt.show()
```

```python
r = 1
n = 25
phis = np.linspace(0, 180, num=n)
thetas = np.linspace(0, 360, num=n)
x = r * np.outer(np.sin(phis), np.cos(thetas))
y = r * np.outer(np.sin(phis), np.sin(thetas))
z = r * np.outer(np.cos(phis), np.ones(n))
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(x, y, z)
plt.show()
```

```python
data = np.concatenate((x.reshape(-1, 1), y.reshape(-1,1), z.reshape(-1,1)), axis=1)
data.shape
```

```python
pca = PCA()
```

```python
pca.fit(data)
```
