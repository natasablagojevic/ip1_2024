```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.metrics import confusion_matrix
```

```python
data = pd.read_csv('./datasets/data.csv')
```

```python
data.head()
```

```python
data.isna().any().any()
```

```python
data.isna().any()
```

```python
data.isna().sum()
```

```python
data.shape
```

```python
data = data.drop('id', axis=1)
```

```python
data
```

```python
data = data.drop('Unnamed: 32', axis=1)
```

```python
data
```

```python
Y = data['diagnosis']
X = data.drop('diagnosis', axis=1)
feature_names = X.columns
```

```python
Y.hist()
```

```python
Y.replace('B', 0, inplace=True)
Y.replace('M', 1, inplace=True)
```

```python
Y.unique()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=123, stratify=Y)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_train = pd.DataFrame(X_train, columns=feature_names)
```

```python
X_train.head()
```

```python
X_train.describe()
```

```python
model = KNeighborsClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_train)
```

```python
print(pd.DataFrame(confusion_matrix(y_train, y_pred), columns=['B', 'M'], index=['B', 'M']))
```

```python
y_test_pred = model.predict(X_test)
```

```python
print(pd.DataFrame(confusion_matrix(y_test, y_test_pred), columns=['B', 'M'], index=['B', 'M']))
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_train, y_pred))
```

```python
print(classification_report(y_test, y_test_pred))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {
    'n_neighbors': range(10, 50, 5),
    'weights' : ['uniform', 'distance'],
    'p' : [1, 2]
}
```

```python
estimator = GridSearchCV(KNeighborsClassifier(), param_grid=params, verbose=2)
```

```python
estimator.fit(X_train, y_train)
```

```python
estimator.best_params_
```

```python
best_model = estimator.best_estimator_
```

```python
y_pred1 = best_model.predict(X_train)
```

```python
y_test_pred1 = best_model.predict(X_test)
```

```python
print(pd.DataFrame(confusion_matrix(y_train, y_pred1), columns=['B', 'M'], index=['B', 'M']))
```

```python
print(classification_report(y_train, y_pred1))
```

```python
print(pd.DataFrame(confusion_matrix(y_test, y_test_pred1), columns=['B', 'M'], index=['B', 'M']))
```

```python
print(classification_report(y_test, y_test_pred1, zero_division=1.0))
```
