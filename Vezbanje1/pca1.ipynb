```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
```

```python
data
```

```python
X = data['data']
```

```python
Y = data['target']
```

```python
feature_names = data['feature_names']
```

```python
class_names = data['target_names']
```

```python
X
```

```python
X = pd.DataFrame(X, columns=feature_names)
```

```python
X
```

```python
class_names
```

```python
Y
# 0 -> Benign
# 1 -> Malignant
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=123, stratify=Y)
```

```python
X.shape
```

```python
Y.shape
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(X_train)
```

```python
X_train_pca = pca.transform(X_train)
```

```python
X_test_pca = pca.transform(X_test)
```

```python
pca.n_components_
```

```python
pca_names = [f'pca_{i}' for i in range(pca.n_components_)]
```

```python
pca_names
```

```python
plt.matshow(pca.components_)
plt.xticks(ticks=range(pca.components_.shape[1]), labels=feature_names, rotation='vertical')
plt.yticks(ticks=range(pca.n_components_), labels=pca_names)
plt.colorbar()
plt.show()
```

```python
# objasnjena varijansa
pca.explained_variance_
```

```python
# udeo objasnjene varijanse 
pca.explained_variance_ratio_
```

```python
plt.bar(height=pca.explained_variance_ratio_, x=pca_names)
plt.title('Udeo objasnjene varijanse')
plt.show()
```

```python
scatter = plt.scatter(X_train_pca[:, 0], X_train_pca[:, 1], c=y_train)
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.legend(handles=scatter.legend_elements()[0], labels=class_names.tolist())
plt.show()
```
