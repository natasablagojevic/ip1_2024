```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv('../datasets/data.csv')
```

```python
data
```

```python
data.isna().any()
```

```python
data.isna().sum()
```

```python
data.shape
```

```python
data.drop('Unnamed: 32', axis=1, inplace=True)
```

```python
data.drop('id', axis=1, inplace=True)
```

```python
data
```

```python
Y = data['diagnosis']
X = data.drop('diagnosis', axis=1)
```

```python
feature_names = X.columns
```

```python
Y.unique()
```

```python
Y.replace('B', 0, inplace=True)
Y.replace('M', 1, inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=123, stratify=Y)
```

```python
data
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
def report(y, y_pred):
    print(classification_report(y, y_pred))
    print('-'*100)
    print(pd.DataFrame(confusion_matrix(y, y_pred), columns=['B', 'M'], index=['B', 'M']))
```

```python
report(y_train, model.predict(X_train))
```

```python
report(y_test, model.predict(X_test))
```

```python
params = {
    'max_depth': [2,4,6,8,10],
    'criterion': ['gini', 'entropy']
}
```

```python
estimator = GridSearchCV(DecisionTreeClassifier(), param_grid=params, cv=5, verbose=10)
```

```python
estimator.fit(X_train, y_train)
```

```python
estimator.best_estimator_
```

```python
model = estimator.best_estimator_
```

```python
report(y_train, model.predict(X_train))
```

```python
report(y_test, model.predict(X_test))
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(12,8))
plot_tree(model, feature_names=feature_names, filled=True)
plt.title(f'Stabla Odlucivanja')
plt.show()
```
