```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../datasets/data.csv')
```

```python
data
```

```python
data.isna().any().any()
```

```python
data.isna().any()
```

```python
data.isna().sum()
```

```python
data.shape
```

```python
data.drop('Unnamed: 32', axis=1, inplace=True)
```

```python
data.drop('id', axis=1, inplace=True)
```

```python
data.isna().any().any()
```

```python
from sklearn.model_selection import train_test_split
```

```python
Y = data['diagnosis']
X = data.drop('diagnosis', axis=1)
```

```python
feature_nemas = X.columns
```

```python
Y.hist()
```

```python
edges, lines, bins = plt.hist(Y)
plt.bar_label(bins)
plt.show()
```

```python
Y.replace('B', 0, inplace=True)
Y.replace('M', 1, inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=123, stratify=Y)
```

```python
X_train.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from sklearn.metrics import confusion_matrix, classification_report
```

```python
def report(y, y_pred):
    print(classification_report(y, y_pred))
    print('-'*80)
    print(pd.DataFrame(confusion_matrix(y, y_pred), index=['B', 'M'], columns=['B', 'M']))
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
y_train_pred = model.predict(X_train)
```

```python
report(y_train, y_train_pred)
```

```python
y_test_pred = model.predict(X_test)
```

```python
report(y_test, y_test_pred)
```

```python
model.get_params()
```

```python
plot_tree(model, feature_names=feature_nemas, filled=True)
plt.title(f'DecisionTreeClassifier with deafult setup')
plt.show()
```

```python
model = DecisionTreeClassifier(criterion='gini', max_depth=5)
```

```python
model.fit(X_train, y_train)
```

```python
y_train_pred = model.predict(X_train)
```

```python
report(y_train, y_train_pred)
```

```python
y_test_pred = model.predict(X_test)
```

```python
report(y_test, y_test_pred)
```

```python
plot_tree(model, feature_names=feature_nemas, filled=True)
plt.title(f'DecisionTreeClassifier with max depth 5')
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [2,4,6,8]
}
```

```python
estimator = GridSearchCV(DecisionTreeClassifier(), param_grid=params, cv=3, verbose=3)
```

```python
estimator.fit(X_train, y_train)
```

```python
estimator.best_estimator_
```

```python
model = estimator.best_estimator_
```

```python
report(y_train, model.predict(X_train))
```

```python
report(y_test, model.predict(X_test))
```

```python
plot_tree(model, feature_names=feature_nemas, filled=True)
plt.title(f'DecisionTreeClassifier with depth: {model.get_depth()}')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(criterion='gini', max_depth=4)
```

```python
model.get_params()
```

```python
model.fit(X_train, y_train)
```

```python
report(y_train, model.predict(X_train))
```

```python
report(y_test, model.predict(X_test))
```

```python
plot_tree(model.estimators_[0], feature_names=feature_nemas, class_names=['B', 'M'] ,filled=True)
plt.title('RandomForestClassifier')
plt.show()
```

```python
estimator = GridSearchCV(RandomForestClassifier(), param_grid=params, cv=3, verbose=3)
```

```python
estimator.fit(X_train, y_train)
```

```python
estimator.best_estimator_
```

```python
model = estimator.best_estimator_
```

```python
report(y_train, model.predict(X_train))
```

```python
report(y_test, model.predict(X_test))
```

```python
plot_tree(model.estimators_[0], class_names=['B', 'M'], feature_names=feature_nemas, filled=True)
plt.title('RandomForestClassifier')
plt.show()
```
